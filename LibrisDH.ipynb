```
import requests
import json
r = requests.get("http://libris.kb.se/xsearch?query=TIT:(digital*+humani*)&format=json&n=200")
data = json.loads(r.content)

list = data['xsearch']['list']
```

```
for book in list:
  print(book.get('title'))
  print(book.get('creator'))
  print(book.get('date'))
  
  print("--------------------------------------")
```
